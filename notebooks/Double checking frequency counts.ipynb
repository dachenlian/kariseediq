```python
from collections import Counter
from itertools import zip_longest
import re

import pandas as pd
```

```python
df = pd.read_csv('seediq_items_updated-20191201-sung-meaning-20191114_sentences-only.csv', header=None, names=['sent'])
```

```python
sent_boundary = ['.', '。', '!', '！', '?', '？', ';', '；']
SENT_BOUNDARY_RE = re.compile(rf'[{"".join(sent_boundary)}]')
punctuations = "".join(['，', '。', '？', '！', '：', '；', '（', '）', '「', '』',
                        ',', '.', '?', '!', ':', ';', '(', ')', '"', '“',
                        '”', '~', '/', '-'])  # 共17個
PUNCTUATION_RE = re.compile(rf'[{"".join(punctuations)}]')
```

```python
def clean(row):
    row['sent'] = row['sent'].replace('\n', '')
    return row
df = df.apply(clean, axis=1)
```

```python
def split_sent(sents):
    sent = []
    for s in sents:
        r = list(filter(not_empty, SENT_BOUNDARY_RE.split(s)))
#         if len(r) > 1:
#             print(len(r))
        sent.extend(list(filter(not_empty, r)))
    return sent
def split_sent_v2(sents):
    return list(filter(not_empty, SENT_BOUNDARY_RE.split(sents)))
```

```python
def not_empty(s):
    if not s:
        return False
    elif s.isspace():
        return False
    elif s == 'NULL':
        return False
    return True
```
