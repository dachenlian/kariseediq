# KWIC

```python
def _load_texts():
    with open('/home/richard/Downloads/kwic3_20170913/seediq_corpus/dict_sentences.txt') as f:
        texts = f.readlines()
    return " ".join(texts)


def build_kwic(query, width, include_examples=False):
    # add space between char and punctuation
    pat_one = r"(\w+)([{}])".format(string.punctuation)
    # add space between punctuation and char
    pat_two = r"([{}])(\w+)".format(string.punctuation)

    texts = _load_texts()
    texts = re.sub('\n', ' ', texts).strip()
    texts = re.sub(pat_one, r'\1 \2', texts)
    texts = re.sub(pat_two, r'\1 \2', texts)
    word_gen = (word for word in texts.split())
    text = Text(word_gen)
    conc_list = text.concordance_list(
        query, width=width, lines=9999999)  # show all lines
    return conc_list
```

```python
kwic = build_kwic('ka', 80)
```

```python
def sort_kwic(k, side='left', window=2):
    if side == 'left':
        k.sort(key=lambda line: [w.lower() for w in line.left[-window:]])
    else:
        k.sort(key=lambda line: [w.lower() for w in line.right[:window]])
    return k
```

```python
sorted_k = sort_kwic(kwic, side='right', window=1)
```

```python
for k in sorted_k:
    print(k.line)
```

```python
import nltk
from nltk.collocations import *
```

```python
with open('/home/richard/Downloads/kwic3_20170913/seediq_corpus/dict_sentences.txt') as f:
    words = f.read().split()

bigram_measures = nltk.collocations.BigramAssocMeasures()
trigram_measures = nltk.collocations.TrigramAssocMeasures()
finder = BigramCollocationFinder.from_words(words)
```

```python
finder.apply_freq_filter(3)
finder.nbest(bigram_measures.pmi, 10)
```
